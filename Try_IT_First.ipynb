```python
# Importing pandas for data processing task
import pandas as pd
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Assignment_TRY_IT_FIRST/Flipkart.csv')
df1.head()
```

```python
# Checking columns available in dataset
df1.columns
```

```python
# Converting columns to corresponding rows
df_melt=pd.melt(df1, id_vars=['Sr No','Product Name'],value_vars=['Flipkart Link - Clr 1',
       'Flipkart Link - Clr 2', 'Flipkart Link - Clr 3',
       'Flipkart Link - Clr 4', 'Flipkart Link - Clr 5',
       'Flipkart Link - RAM1', 'Flipkart Link - RAM2', 'Flipkart Link - RAM3',
       'Flipkart Link - RAM4', 'Flipkart Link - Storage 1',
       'Flipkart Link - Storage 2', 'Flipkart Link - Storage 3',
       'Flipkart Link - Storage 4'],var_name='Flipkart Link',value_name='Link')
df_melt
```

```python
# Droping the NAN Rows 
df_melt = df_melt.dropna()
```

```python
df_melt
```

```python
# Checking for duplicate links in dataset
df_melt.Link.duplicated()
```

```python
# Total 4 duplicates are there in dataset
df_melt.Link.duplicated().sum()
```

```python
# Removed duplicates Links
df2 = df_melt.drop_duplicates(subset=["Link"],keep=False)
```

```python

df2.Link.duplicated().sum()
```

```python
df2
```

```python
# Shorten the URL 
shorten_url = []
for i in df2['Link']:
  re = i.split('&lid')
  res = re[0]
  shorten_url.append(res)
  print(res)
  
```

```python
shorten_url
```

```python
# List added to Dataset
df2['link'] = shorten_url
```

```python
df2
```

```python
# Removed redundant Link Column 
df2.drop(['Link'],axis=1)
```
